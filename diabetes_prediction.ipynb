```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
db_data = pd.read_csv('diabetes.csv')
db_data.head()
```

```python
db_data.info()
```

```python
db_data.describe()
```

```python
db_data['Outcome'].value_counts()
```

```python
X = db_data.drop(columns = "Outcome", axis =1)
Y = db_data["Outcome"]
```

```python
scaler = StandardScaler()
scaler.fit(X)
std_data = scaler.transform(X)
```

```python
std_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.225,  random_state = 4)
```

```python
model= LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_pred = model.predict(X_train)
train_acc = accuracy_score(X_pred, Y_train)
print(train_acc)
```

```python
Xt_pred = model.predict(X_test)
test_acc = accuracy_score(Xt_pred, Y_test)
print(test_acc)
```
